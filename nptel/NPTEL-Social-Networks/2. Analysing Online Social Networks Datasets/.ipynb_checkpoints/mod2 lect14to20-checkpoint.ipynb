#### gml_ 
    -> Graph Modelling Language
>#### Without attributes
    graph [
        node [
            id a
            label 'Node A'
        ]
        node [
            id b
            label 'Node b'
        ]
        edge [
            source a
            target b
            label 'E1'
        ]
    ]
    <br>
>#### With attributes
    graph [
        heirarchic 1
        directed 1
        node [
            id a
            label 'Node A'
            graphics[
                x 0
                y 0
            ]
        ]
        node [
            id b
            label 'Node b'
            graphics[
                x 20
                y 30
            ]
        ]
        edge [
            source a
            target b
            label 'E1'
            LabelGraphics[
            text 'Happy New Year'
            ]
        ]
    ]
